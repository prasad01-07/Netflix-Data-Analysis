```python
df['director'].fillna('Unknown', inplace=True)

df['cast'].fillna('Not Available', inplace=True)

df['country'].fillna('Unknown', inplace=True)

df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)

df['rating'].fillna(df['rating'].mode()[0], inplace=True)

df['duration'].fillna(df['duration'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

Data Cleaning Steps

1. Missing values in director were replaced with "Unknown".

2. Missing values in cast were replaced with "Not Available".

3. Missing values in country were replaced with "Unknown".

4. Missing values in date_added, rating, and duration were replaced using the most frequent value (mode).

5. Data was checked for duplicate records and cleaned for analysis.

Question 1: How many Movies and TV Shows are available on Netflix?

```python
df['type'].value_counts()
```

```python
df['type'].value_counts()
```

Insight:
Netflix contains significantly more Movies (6131) than TV Shows (2676), indicating that movies dominate the platform's content library.

Question 2: Which countries have the highest number of Netflix titles?

```python
df['country'].value_counts().head(10)
```

Insight:
The United States has the highest number of Netflix titles (2818), followed by India (972) and the United Kingdom (419). This shows that Netflix content is heavily concentrated in these countries.

Question 3: What are the most common content ratings on Netflix?

```python
df['rating'].value_counts()
```

Insight:
TV-MA is the most common rating on Netflix with 3207 titles, followed by TV-14 with 2160 titles. This suggests that Netflix primarily offers content targeted at mature and teenage audiences.

Question 4: Which release years have the highest number of Netflix titles?

```python
df['release_year'].value_counts().head(10)
```

Insight:
Netflix added the highest number of titles released in 2018 (1147 titles). The years 2017, 2019, and 2020 also contributed significantly, indicating rapid content growth during this period.

Question 5: Which directors have the most titles on Netflix?

```python
df['director'].value_counts().head(10)
```

Insight:
Rajiv Chilaka has the highest number of titles on Netflix (19), followed by Raúl Campos & Jan Suter (18). This indicates that these directors have made significant contributions to Netflix's content library.

Visualization 1: Movies vs TV Shows

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')

plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')

plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Countries by Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')

plt.show()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')

plt.title('Top Netflix Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

plt.show()
```
